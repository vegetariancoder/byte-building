#### 1. Customer Lifetime Value (CLV)
- Aggregate total spend per customer.

```python
import dlt
from pyspark.sql.functions import col, current_date, sum
```

```python
@dlt.view(
    name="vw_customer_lifetime_value",
    comment="Customer lifetime value (CLV)"
)
def customer_lifetime_value():
    cur_customer_order_history = dlt.read_stream("cur_customer_order_history")

    cur_customer_order_history_df = cur_customer_order_history\
        .groupBy("customer_id")\
        .agg(sum("item_price").alias("total_spend"))\
        .withColumn("load_date", current_date())

    return cur_customer_order_history_df
```
